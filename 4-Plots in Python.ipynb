```python
# From matplotlib, import pyplot under the alias plt
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# para abrir debemos de tener cuidado el tipo de separacion 
# abrimos y vemos si se ha cargado correctamente
cellphone = pd.read_csv('cellphone.csv',sep = ';')
cellphone.head()
```

```python
# Create a scatter plot of the data from the DataFrame cellphone
plt.scatter(cellphone.x,cellphone.y)

# Add labels
plt.ylabel('Latitude')
plt.xlabel('Longitude')

# Display the plot
plt.show()
```

```python
# Change the marker color to red
plt.scatter(cellphone.x, cellphone.y,
           color='red')

# Add labels
plt.ylabel('Latitude')
plt.xlabel('Longitude')

# Display the plot
plt.show()
```

```python
# Change the marker shape to square
plt.scatter(cellphone.x, cellphone.y,
           color='red',
           marker='s')

# Add labels
plt.ylabel('Latitude')
plt.xlabel('Longitude')

# Display the plot
plt.show()
```

```python
# Change the transparency to 0.1
plt.scatter(cellphone.x, cellphone.y,
           color='red',
           marker='s',
           alpha=0.1)

# Add labels
plt.ylabel('Latitude')
plt.xlabel('Longitude')

# Display the plot
plt.show()
```

```python
# para abrir debemos de tener cuidado el tipo de separacion 
# abrimos y vemos si se ha cargado correctamente
puppies = pd.read_csv('puppies.csv',sep = ';')

puppies.head()
```

```python
# Create a histogram of the column weight
# from the DataFrame puppies
plt.hist(puppies.weight)

# Add labels
plt.xlabel('Puppy Weight (lbs)')
plt.ylabel('Number of Puppies')

# Display
plt.show()
```

```python
# Change the number of bins to 50
plt.hist(puppies.weight,
        bins=50)

# Add labels
plt.xlabel('Puppy Weight (lbs)')
plt.ylabel('Number of Puppies')

# Display
plt.show()
```
